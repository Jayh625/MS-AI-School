## 인공신경망 가중치 적용한 시각화 실습

```python
import torch
import matplotlib.pyplot as plt
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
# 2차원 가중치 행렬을 랜덤으로 초기화
w = torch.randn(3,3)
print(w)

w_flattened = w.flatten()
print(w_flattened)
```

### 시각화

```python
plt.hist(w_flattened, bins=20)
plt.xlabel("Weight value")
plt.ylabel("Frequency")
plt.show()
```

## 크기가 1인 랜덤한 값을 가지는 편향을 생성

```python
b =  torch.rand(1) # 편향을 랜덤으로 초기화

# 편향 값을 시각화
plt.hist(b.detach().numpy(), bins=20) # b.detach().numpy() -> 텐서를 넘파이 배열로 변환
plt.xlabel("Bias value")
plt.ylabel("Frequency")
plt.show()
```
