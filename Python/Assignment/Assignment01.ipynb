### 1. 먼저 가상의 데이터를 생성해보겠습니다. 
### 이 데이터는 1학년부터 3학년까지의 학생들의 수학, 영어, 국어점수를 나타내며, 각 학년당 100명씩의 학생이 있다고 가정합니다.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import random
import os
np.random.seed(42)
```

```python
def scale_score(score, min_score, max_score):
    if score < 0 :
        score = abs(score)
    scaled_score = int((score - min_score) / (max_score - min_score) * 100)
    return scaled_score
def create_student_score_dataset(grade, std_num, sub1, sub2, sub3):
    scores = []
    for grade in range(1, grade+1):
        for _ in range(std_num):
            math_score = scale_score(np.random.normal(loc=50, scale=10), 0, 100)
            eng_score = scale_score(np.random.normal(loc=60, scale=10), 0, 100)
            kor_score = scale_score(np.random.normal(loc=70, scale=10), 0, 100)

            scores.append((grade, math_score, eng_score, kor_score))
    df = pd.DataFrame(scores, columns=['Grade', sub1, sub2, sub3])
    return df

df = create_student_score_dataset(3, 100, 'Math', 'English', 'Korean')
df
```

### 2. CSV 파일로저장 ex) student_scores.csv

```python
# 폴더 생성
os.makedirs("./data", exist_ok=True)

# csv 파일로 저장
df.to_csv('./data/student_scores.csv')
```

### 3. 상관관계 분석 - 히트맵 출력

```python
def draw_heatmap(data, *args):
    cols = list(args)
    corr = data[cols].corr()
    plt.figure(figsize=(6,6))
    sns.heatmap(corr, annot=True, cmap='YlGnBu')
    plt.title("Correlation between subjects")
    plt.show()
draw_heatmap(df, 'Math', 'English', 'Korean')
```
