### 단순 선형 회귀의 개념 간단 실습 - 최소 제곱법을 이용한 단순 선형 회귀 구현

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 생성
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])
```

```python
# 산점도 그리기 
plt.scatter(x,y)
plt.show()
```

```python
beta_1 = np.sum((x - np.mean(x)) * (y - np.mean(y))) / np.sum((x - np.mean(x))**2)
beta_0 = np.mean(y) - beta_1 * np.mean(x)
```

```python
plt.scatter(x,y)
plt.plot(x, beta_0 + beta_1 * x, color='red')
plt.show()
```
