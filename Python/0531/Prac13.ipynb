```python
import pandas as pd
import numpy as np
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_acf
```

```python
n = 100 # 데이터 포인트 수
ar_params = [0.7] # 자기 회귀 계수
ma_params = [] # 이동 평균 계수
residuals = np.random.normal(0,1,n) # 전차 생성

# AR(1) 모델 생성
data = [0] # 초기값 설정
for i in range(1,n) : 
    data.append(ar_params[0] * data[i-1] + residuals[i])

# 데이터 프레임 생성
df = pd.DataFrame({
    'Date' : pd.date_range(start='2023-01-01', periods=n, freq='D'),
    'Value' : data
})
df.head()
```

```python
# AR(1) 모델 적합
model = AutoReg(df['Value'], lags=1)
results = model.fit()

# 모델 정보 출력 
print(results.summary())

# ACF 그래프 출력
plot_acf(df['Value'], lags=20)
```
