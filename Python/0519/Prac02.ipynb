```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 정규분포를 따르는 데이터 생성
data = np.random.normal(loc=0, scale=1, size=10000)

# 데이터 분포 시각화
sns.histplot(data)
plt.show()
# 데이터 통계 지표 계산 
mean = np.mean(data)
std = np.std(data)
print("평균 :", mean)
print("표준편차 :", std)
```
