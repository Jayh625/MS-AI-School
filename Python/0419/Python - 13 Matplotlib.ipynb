```python
# plt.scatter 산점도를 표현

# rand (0~1)의 값을 지정된 숫자 만큼 임의의 값을 만들어 낸다. 
# randn (0)을 기준으로 편차만큼 떨어진 임의값을 지정된 숫자만큼 만들어 낸다. 
# randint 지정된 범위안의 정수 1개를 만들어 낸다. 
# x = rng.randn(100)  # 0을 기준으로해서 +,- 랜덤한 값들  
# y = rng.rand(100)   # 0.몇 랜덤한 값들
# z = np.random.randint(1,9) # 1~9 임의의 값 1개 

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
color = rng.rand(100)
sizes = rng.rand(100) * 1000
plt.scatter(x,y, s=sizes, alpha=0.3, c=color, cmap='viridis')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
features.shape
```

```python
# 데이터가 많이 겹치면(색깔이 진할수록) 노이즈
plt.style.use('default')
plt.scatter(features[2],features[3],c=iris.target, s=features[3]*100, alpha=0.2, cmap='viridis') 
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
print(iris.feature_names[3])
```

```python
# 오차 시각화 하기
x = np.linspace(0,10,50)
dy = 0.8 # 오차범위
y = np.sin(x + np.random.randn(50))
plt.errorbar(x,y,yerr=dy,fmt='.k')
```

```python
plt.errorbar(x,y, yerr=dy, fmt='o', color='black',ecolor='lightgrey',elinewidth=3)
plt.title('Error Chart')
```
