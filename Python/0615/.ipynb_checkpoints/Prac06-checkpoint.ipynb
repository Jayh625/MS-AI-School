```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

#   
image = cv2.imread('./data/image02.jpeg')
# image BGR2RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#   
# 1 , 0 , -1 
flipped1 = cv2.flip(image,1)
flipped2 = cv2.flip(image,0)
flipped3 = cv2.flip(image,-1)

#   
plt.imshow(image)
plt.show()

#    
plt.imshow(flipped1)
plt.show()

#    
plt.imshow(flipped2)
plt.show()

#    
plt.imshow(flipped3)
plt.show()
```
