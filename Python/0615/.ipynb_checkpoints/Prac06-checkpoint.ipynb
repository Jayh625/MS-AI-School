```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 이미지 파일 불러오기
image = cv2.imread('./data/image02.jpeg')
# image BGR2RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 이미지 좌우 반전
# 1 좌우, 0 상하, -1 좌우상하
flipped1 = cv2.flip(image,1)
flipped2 = cv2.flip(image,0)
flipped3 = cv2.flip(image,-1)

# 원본 이미지 보기
plt.imshow(image)
plt.show()

# 이미지 좌우 반전 
plt.imshow(flipped1)
plt.show()

# 이미지 상하 반전 
plt.imshow(flipped2)
plt.show()

# 이미지 좌우상하 반전 
plt.imshow(flipped3)
plt.show()
```
