```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 이미지 파일 불러오기
image = cv2.imread('./data/image02.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 색상 보정
org_img = image.copy()
balance = [0.8, 0.7, 0.8]

for i, value in enumerate(balance) :
    if value != 1.0 :
        org_img[:,:,i] = cv2.addWeighted(org_img[:,:,i], value, 0, 0, 0)
        # cv2.addWeighted : src에 대해 value 만큼의 가중치로 색온도 조절

# 원본 이미지
plt.imshow(image)
plt.show()

# 색보정 이미지
plt.imshow(org_img)
plt.show()
```
