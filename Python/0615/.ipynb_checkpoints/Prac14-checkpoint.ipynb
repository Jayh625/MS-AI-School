```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 이미지 파일 불러오기
image = cv2.imread('./data/image02.jpeg')
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 적응형 임계값 처리
thresh = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
# ADAPTIVE_THRESH_MEAN_C : 적응형 임계값 처리, 임계값 기준을 평균치를 사용함
# 인자 11 : 블록 크기, 11x11 블록으로 이미지를 나눔

# 원본 이미지 
plt.imshow(image, 'gray')
plt.show()

# 이미지 블러링 적용
plt.imshow(thresh, 'gray')
plt.show()
```
