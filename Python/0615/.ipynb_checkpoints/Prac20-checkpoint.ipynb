```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 히스토그램 균일화 적용
img = cv2.imread('./data/image02.jpeg', cv2.IMREAD_GRAYSCALE)

# cv2.equalizeHist 함수가 8비트 단일 채널 이미지를 입력 -> 그레이 이미지를 사용
img_equalized = cv2.equalizeHist(img)

# 원본 이미지 
plt.imshow(img, 'gray')
plt.show()

# 히스토그램 균일화 생성 이미지
plt.imshow(img_equalized, 'gray')
plt.show()
```
