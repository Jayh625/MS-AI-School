```python
import cv2
import matplotlib.pyplot as plt
```

```python
# image read
image = cv2.imread('./data/image02.jpeg')

# image BGR2RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#   
angle = 30

#      
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, angle, 1.0)

#  
rotated = cv2.warpAffine(image, M, (w,h))

#   
plt.imshow(image)
plt.show()

# rotated  
plt.imshow(rotated)
plt.show()
```
