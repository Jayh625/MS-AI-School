```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# 이미지 파일 불러오기
image = cv2.imread('./data/image02.jpeg')
# image BGR2RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Salt-and-Pepper Noise 추가
noise = np.zeros(image.shape, np.uint8) # (h, w, c),  c -> color channel : B, G, R
cv2.randu(noise, 0, 255)
black = noise < 30
white = noise > 225
noise[black] = 0
noise[white] = 255

# 각 채널 별로 노이즈 추가
noise_b = noise[:, :, 0]
noise_g = noise[:, :, 1]
noise_r = noise[:, :, 2]
noisy_img = cv2.merge([cv2.add(image[:,:,0], noise_b),
                       cv2.add(image[:,:,1], noise_g),
                       cv2.add(image[:,:,2], noise_r)])

# 원본 이미지 보기
plt.imshow(image)
plt.show()

# Noise 추가된 이미지 출력
plt.imshow(noisy_img)
plt.show()
```
