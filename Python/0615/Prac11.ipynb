```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#   
image = cv2.imread('./data/image02.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#     
x, y, w, h = 100, 100, 200, 200

#  
crop_img = image[y:y+h, x:x+w]

#  
plt.imshow(image)
plt.show()

#  
plt.imshow(crop_img)
plt.show()
```
