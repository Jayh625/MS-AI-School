```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
math_scores = np.random.normal(loc=70, scale=10, size=(300,))
english_scores = np.random.normal(loc=80, scale=10, size=(300,))
korean_scores = np.random.normal(loc=75, scale=10, size=(300,))
```

```python
grades = np.repeat([1,2,3], 100)
```

```python
df = pd.DataFrame({
    'Grade' : grades,
    'Math' : math_scores,
    'English' : english_scores,
    'Korean' : korean_scores
})
df.to_csv('./data/student_score.csv', index=False)
```

```python
df = pd.read_csv('./data/student_score.csv')
df
```

```python
plt.hist(df['Math'], bins=10)
plt.show()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```
