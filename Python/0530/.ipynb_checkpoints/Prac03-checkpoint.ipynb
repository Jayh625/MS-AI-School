## 시계열 분해 : 추세, 계절성, 주기, 불규칙성 요소 분리하여 시각화

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

### 시계열 데이터 생성

```python
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100)

# 데이터 프레임 생성
data = pd.DataFrame({
    'Date' : dates,
    'Value' : values
})
data.head()
```

```python
# 시계열 분해
result = seasonal_decompose(data['Value'], model='additive', period=7)

# 분해 결과 시각화
plt.figure(figsize=(10,8))

# 원본 데이터
plt.subplot(4,1,1)
plt.plot(data['Date'], data['Value'])
plt.title('Original Data')
plt.ylabel('Value')

plt.subplot(4,1,2)
plt.plot(data['Date'], result.trend)
plt.title('Trend')
plt.ylabel('Trend')

plt.subplot(4,1,3)
plt.plot(data['Date'], result.seasonal)
plt.title('Seasonality')
plt.ylabel('Seasonality')

plt.subplot(4,1,4)
plt.plot(data['Date'], result.resid)
plt.title('Irregularity')
plt.ylabel('Irregularity')

plt.tight_layout()
plt.show()
```
