# 시계열 데이터 구성 요소 3가지(상향, 하향, 평평한 추세) 시각화

- 추세 (Trend) : 데이터가 일정한 방향으로 증가하거나 감소하는 경향
- 계절성 (Seasonality) : 데이터에 주기적인 패턴이 존대
- 주기 (Cycle) : 데이터에 장기적인 주기성이 관찰
- 불규칙성 (Irregularity) : 데이터에 예측하기 어려운 불규칙한 변동이 존재하는 경우

- 분산(Variance): 데이터 값들의 퍼짐도 를 측정하는 지표로, 평균값과의 거리의 제곱의 평균입니다. 분산이 클수록 데이터 값들이 평균에서 멀리 퍼져 있는 것을 나타냅니다.
- 표준편차(Standard Deviation): 분산의 양의 제곱근으로, 데이터 값들의 퍼짐정도를 나타내는 지표입니다. 표준편차가 크면 데이터 값들이 평균에서 상대적으로 더 멀리 퍼져 있는 것을 의미합니다.
- 변동계수(Coefficient of Variation): 표준편차를 평균으로 나눈 값으로, 상대적인 변동의 정도를 나타내는 지표입니다. 변동계수가 크면 상대적으로 큰 변동성을 가진 것으로 해석됩니다

### 데이터 생성 (랜덤하게 시계열 데이터 생성)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100).cumsum()

# 데이터 프레임 생성
data = pd.DataFrame({
    'Date' : dates,
    'Value' : values
})
data
```

### 상향 추세 시각화

```python
data['Upward_Trend'] = np.linspace(0, 50, len(data))
data
```

```python
plt.plot(data['Date'], data['Upward_Trend'], label='Upward_Trend')
plt.legend()
plt.xticks(rotation=-45)
plt.show()
```

### 하향 추세 시각화

```python
data['Downward_Trend'] = np.linspace(50, 0, len(data))
data
```

```python
plt.plot(data['Date'], data['Downward_Trend'], label='Downward_Trend')
plt.legend()
plt.xticks(rotation=-45)
plt.show()
```

### 평평한 추세 시각화

```python
data['Flat_Trend'] = np.full(len(data), 50)
data
```

```python
plt.plot(data['Date'], data['Flat_Trend'], label='Flat_Trend')
plt.legend()
plt.xticks(rotation=-45)
plt.show()
```
