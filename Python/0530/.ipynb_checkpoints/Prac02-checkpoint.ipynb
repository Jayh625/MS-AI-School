# 시계열 데이터 탐색 및 시각화
- 시계열 데이터 시각화 소개 (선 그래프, 박스 플롯, 자기상관 그래프)

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

# random seed 고정
np.random.seed(42)

# row 출력제한 해제
pd.set_option('display.max_row', 1000)

# column 출력제한 해제
pd.set_option('display.max_columns', 100)
```

```python
# 시계열 데이터 생성
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)

# 데이터 프레임 생성
data = pd.DataFrame({
    'Date' : dates,
    'Value' : values
})

# 시계열 데이터 시각화 
plt.plot(data['Date'], data['Value'])
plt.xticks(rotation=-45)
plt.show()
```

### 산점도 그래프

```python
# 시계열 데이터 생성
date_temp = pd.date_range(start='2023-01-01', periods=1000, freq='D')
val_temp = np.random.randn(1000)

# 프레임 생성
data_pd = pd.DataFrame({
    'Date' : date_temp,
    'Value' : val_temp
})
data_pd.head()
```

### 산점도 그래프 시각화

```python
plt.scatter(data_pd['Date'], data_pd['Value'])
plt.xticks(rotation=-45)
plt.show()
```

### 막대 그래프 시각화

```python
dates_bar_plot = pd.date_range(start='2023-01-01', periods=1000, freq='D')
val_bar_plot = np.random.randn(1000)

bar_plot_data = pd.DataFrame({
    'Date' : dates_bar_plot,
    'Value' : val_bar_plot
})
bar_plot_data.head()
```

```python
plt.bar(bar_plot_data['Date'], bar_plot_data['Value'])
plt.xticks(rotation=-45)
plt.show()
```

### 박스 플롯 시각화

```python
box_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
box_val = np.random.randn(1000)

box_data = pd.DataFrame({
    'Date' : box_date,
    'Value' : box_val
})
box_data.head()
```

```python
plt.boxplot(box_data['Value'])
plt.show()
```

### 히스토그램 시각화

```python
histogram_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
histogram_val = np.random.randn(1000)

histogram_data = pd.DataFrame({
    'Date' : histogram_date,
    'Value' : histogram_val
})
histogram_data.head()
```

```python
plt.hist(histogram_data['Value'], bins=20)
plt.show()
```

### 히트맵 시각화

```python
heat_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
heat_val = np.random.randn(1000)

heat_data = pd.DataFrame({
    'Date' : heat_date,
    'Value' : heat_val
})
heat_data.head()
```

```python
# 데이터를 피벗 테이블 변환
pivot_table = heat_data.pivot(index='Date', columns='Value', values='Value')
plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, cmap='YlOrRd', cbar=True)
plt.show()
```

피벗 테이블은 데이터를 특정 열을 기준으로 그룹화하여 요약하는 방법입니다.

Date        Value <br>
23-01-01      A <br>
23-01-01      B <br>
23-01-02      A <br>
23-01-02      C <br>
23-01-03      B <br>
23-01-03      C <br>

피벗 테이블 생성 <br>
             A   B    C <br>
2023-01-01   A   B   NAN <br>
2023-01-02   A   NAN  C <br>
2023-01-03  NAN  B    C <br>
