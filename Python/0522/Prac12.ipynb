```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips_data = sns.load_dataset('tips')
tips_data.head()
```

```python
corr = tips_data[['total_bill', 'tip']].corr()
corr
```

```python
sns.regplot(x='total_bill', y='tip', data=tips_data)
```

```python
# load tips dataset
tips = sns.load_dataset('tips')

# calculate Pearson Correlation coefficient between total_bill and tip
corr = tips['total_bill'].corr(tips_data['tip'], method='pearson')
print("Pearson Correlation Coefficient :", corr)
```

```python
#  
covariance = tips_data['total_bill'].cov(tips_data['tip'])
print("Covariance between total_bill and tips :", covariance)
```
