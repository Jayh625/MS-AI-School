```python
import requests
from io import BytesIO
from PIL import Image
```

```python
img_url = "https://menu.mt.co.kr/moneyweek/thumb/2017/05/17/06/2017051709058028746_1.jpg"
image = Image.open(BytesIO(requests.get(img_url).content))
# response = requests.get(img_url)
# img = BytesIO(response.content)
# image = Image.open(img)
image
```

```python
# 필요한 값들을 준비한다
key = '1e615b74a9fe427798ca9c4ba711e542'
endpoint = 'https://labuser92-computervision.cognitiveservices.azure.com/' + 'vision/v2.0/'
```

```python
ananlyze_endpoint = endpoint + 'analyze'
detect_endpoint = endpoint + 'detect'
```

```python
header = {'Ocp-Apim-Subscription-Key' : key}
params = {'visualFeatures' : 'Categories,Description,Color'}
data = {'url' : img_url}
```

```python
response = requests.post(ananlyze_endpoint, headers=header, params=params, json=data)
```

```python
result = response.json()
result
```
