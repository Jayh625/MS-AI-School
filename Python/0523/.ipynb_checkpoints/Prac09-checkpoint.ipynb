```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2
```

```python
# 모순 설정
df = 5 # 자유도

# 카이제곱 분포에서 랜덤 변수 생성
random_variable = np.random.chisquare(df, size=1000)

plt.hist(random_variable, bins=30, density=True, alpha=0.8)
plt.show()
```
